```python
ax = sns.countplot(x='Your Academic Stage' , data =data)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
plt.figure(figsize=(10,5))
ax = sns.countplot(x='Study Environment' , data =data, palette='rocket')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
plt.figure(figsize=(10,5))
ax = sns.countplot(x='What coping strategy you use as a student?' , data =data, palette='rocket')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
bad_habits_counts = data['Do you have any bad habits like smoking, drinking on a daily basis?'].value_counts(normalize=True)
print(bad_habits_counts)

plt.figure(figsize=(5,4))
plt.pie(bad_habits_counts, labels=bad_habits_counts.index, autopct='%1.1f%%', shadow=True, startangle=90)
plt.show()
```

```python
sns.scatterplot(x='Your Academic Stage', y='Peer pressure' , data=data)
plt.title('Your Academic Stage vs Peer pressure')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='What coping strategy you use as a student?', data=data, palette='viridis')
plt.title('What coping strategy you use as a student? and Academic pressure from your home ')
plt.xlabel('What coping strategy you use as a student?')
plt.ylabel('Academic pressure from your home')
```
